```python
#Print * triangle up to a number of *s
def pyf(r):
    for x in range(r):
        #print(x)
        print(' '*(r-x-1)+'*'*(2*x+1))
        
pyf(10)
```

```python
#Print fibonacci sequence up to a number of terms
input_number = int(input("Enter number of terms required:"))

first_number = 0
second_number = 1

if input_number <=0:
    print("Requested series is ", first_number)

else:
    print(first_number,second_number, end=" ")
    for i in range(2,input_number):
        next= first_number + second_number
        print(next, end=" ")
        first_number = second_number
        second_number = next
```

```python
#Print a list of prime numbers in a given range
from sympy import sieve
def primelist(a,b):
    list= [i for i in sieve.primerange(a,b)]
    return list
primelist(1,100)
```

```python
#Web scraper IMDB top 50 movies, only provide movie name,year and rating

from bs4 import BeautifulSoup
 
import requests
import sys
 
url = "http://www.imdb.com/chart/top"
response = requests.get(url)
soup = BeautifulSoup(response.text)
#print(soup)

tr = soup.findChildren("tr")
#print(tr)

tr = iter(tr)

#print(tr)

next(tr)



for movie in tr:
    rank  = movie.find('td', {'class': 'titleColumn'}).contents[0].strip()
    title = movie.find('td', {'class': 'titleColumn'} ).find('a').contents[0]
    year = movie.find('td', {'class': 'titleColumn'} ).find('span', {'class': 'secondaryInfo'}).contents[0]
    rating = movie.find('td', {'class': 'ratingColumn imdbRating'} ).find('strong').contents[0]
    row = rank + ' ' + title + ' - ' + year + ' ' + ' ' + rating
    
    print(row)
```
