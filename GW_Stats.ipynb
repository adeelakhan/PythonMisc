```python
import pandas as pd
import datetime

#Enter input filename
input_file = 'metrics.log'

#Enter output filename
output_file = 'GW_Stats_16_Sep.csv'

#Add Log filename here to analyze
df = pd.read_csv(input_file,header =None, delim_whitespace=True)
df.dropna(inplace = True)
```

```python
del df[1]
```

```python
df.head(10)
```

```python
headings=["Timestamp", "Prefix", "MAXPIF", "PIF", "MAXPPS"]
df.columns = headings
```

```python
df.head(10)
```

```python
df["Timestamp"] = df["Timestamp"].map(lambda x: x.lstrip(' timestamp='))
df["Prefix"] = df["Prefix"].map(lambda x: x.lstrip(' prefixes='))
df["MAXPIF"] = df["MAXPIF"].map(lambda x: x.lstrip(' packets_in_flight_max='))
df["PIF"] = df["PIF"].map(lambda x: x.lstrip(' packets_in_flight='))
df["MAXPPS"] = df["MAXPPS"].map(lambda x: x.lstrip(' packets_per_second_max='))
df["Timestamp"] = pd.to_datetime(df["Timestamp"], unit = 's').dt.tz_localize('UTC').dt.tz_convert('Australia/Melbourne').dt.strftime('%r')
```

```python
df.head(10)
```

```python
df.to_csv(output_file,index=False)
```
